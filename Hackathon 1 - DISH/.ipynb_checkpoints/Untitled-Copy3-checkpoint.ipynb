```python
Categorical_columns=["Car_Owned",
                     "Bike_Owned",
                     "Active_Loan",
                     "House_Own",
                     "Accompany_Client",
                     "Client_Income_Type",
                     "Client_Education",
                     "Client_Marital_Status",
                     "Client_Gender",
                     "Loan_Contract_Type",
                     "Client_Housing_Type",
                     "Mobile_Tag",
                     "Homephone_Tag",
                     "Workphone_Working",
                     "Client_Occupation",
                     "Cleint_City_Rating",
                     "Application_Process_Day",
                     "Client_Permanent_Match_Tag",
                     "Client_Contact_Work_Tag",
                     "Type_Organization",
                     "Default"]
```

```python
Numerical_columns=[]
for each in data2.columns:
    #print(each)
    if each in Categorical_columns:
        #print(each)
        pass
    else:
        Numerical_columns.append(each)
#######################################################3        
Numerical_columns.remove("ID")
Categorical_columns.remove("Default")
```

```python
#These need some reason why they are not going along.
check=[]
for each in data2.columns:
    if each in Categorical_columns:
        try:
            data2[each]=data2[each].astype("category")
        except:
            check.append(each)
            continue
    elif each in Numerical_columns:
        try:
            data2[each]=data2[each].astype("float64")
        except:
            check.append(each)
            continue          
```

```python
####################################################################
def NaN_treatment(dataframe,Categorical_list, Numerical_list):
    df=dataframe
    j=0
    k=0
    for col in df.columns:
        if col in Numerical_list :
            mean=df[col].mean()
            #print(" Column is ",df[col].dtypes,mean)
            df[col].fillna(value=mean,axis=0,inplace=True)
            k+=1
        elif col in Categorical_list:
            mode=df[col].mode(dropna=True)[0]
            #print(mode)
            #print(" Column is ",df[col].dtypes,mode)
            df[col].fillna(value=mode,axis=0,inplace=True)
            j+=1
        #print("\n")
    print("The number of categorical columns is :", j)
    print("The Number of Numerical columns is :",k)
    return df
#####################################################################
```

```python
data3=data2.copy()
NaN_treatment(data3,Categorical_columns, Numerical_columns) # Can use mean values to replace because we are going to use bagging algorithm
```

```python
data3[Categorical_columns]=data3[Categorical_columns].astype("category")
```
