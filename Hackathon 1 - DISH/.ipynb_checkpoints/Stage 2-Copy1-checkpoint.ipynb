```python
#########################################################3
#########################################################
import multiprocessing
import time
############################################################
#############################################################
def mf(function):
    import time
    multiprocessing_func=function
    starttime = time.time()
    if __name__ == '__main__':
        processes = []
        for i in range(1,10):
            p = multiprocessing.Process(target=multiprocessing_func, args=(i,))
            processes.append(p)
            p.start()    
        for process in processes:
            process.join()
        
    #print()    
    print('Time taken = {} seconds\n'.format(time.time() - starttime))
    return multiprocessing_func
#######################################################################
#####################################################################
```

```python
###################################################################

##################################################################
def value_counts_df_case(dataframe,columns_given): 
    value_counts_out={}
    for each in columns_given:
        value_counts_out.update({each:dataframe[each].value_counts()})
    return value_counts_out
##############################################################

###########################################################
pd.DataFrame(value_counts_df_case(data1,Categorical_columns)).T
```

```python
###########################################################
############################################################
#Function that will list out column by column outliers' Index values. We will check patterns for those.
def outlier_identification(dataframe):
    outliers_values={}
    dataframe_desc=dataframe.describe()
    for col in dataframe.columns:
        Quartile3=list(dataframe[col][dataframe[col]>dataframe_desc[col]["75%"]].index)
        Quartile1=list(dataframe[col][dataframe[col]<dataframe_desc[col]["25%"]].index)
        outliers_values.update({col:[Quartile1,Quartile3]})
    return outliers_values
###########################################################
############################################################
outliers_fault=pd.DataFrame(outlier_identification(data1_default_wo[Numerical_columns]),index=["Q1","Q3"])
outliers_fault
```

```python
outliers_nodefault=pd.DataFrame(outlier_identification(data1_nodefault_wo[Numerical_columns]),index=["Q1","Q3"])
outliers_nodefault
```
