```python
#This is the function I am going to use again and again, so before anything is said
#here is the detail.
#########################################################3
#########################################################
import multiprocessing
import time
############################################################
#############################################################
def mf(function):
    import time
    multiprocessing_func=function
    starttime = time.time()
    if __name__ == '__main__':
        processes = []
        for i in range(1,10):
            p = multiprocessing.Process(target=multiprocessing_func, args=(i,))
            processes.append(p)
            p.start()    
        for process in processes:
            process.join()
        
    #print()    
    print('Time taken = {} seconds\n'.format(time.time() - starttime))
    return multiprocessing_func
#######################################################################
#####################################################################
    
#multiprocessing_func(function) as mf(function) # function() is the function we want multiprocessed
```

```python
############################################################# 
############################################################# 
def remove_wrong_entries(given_Series,d_type_taken):
    Series=given_Series
    d_type=d_type_taken
    for index in range(len(Series)):
        try:
            Series[index]=d_type(Series[index])
        except:
            #print(Series[index])
            Series[index]=np.nan
    return Series
            
#############################################################            
#############################################################             
#data1["Score_Source_3"]=remove_wrong_entries(data1["Score_Source_3"],float).astype("float64")       
#data1["Age_Days"]=remove_wrong_entries(data1["Age_Days"],float).astype("float64")
```

```python
#############################################################            
#############################################################  

def clean_dataframe(dataframe,categorical_list,Numerical_list):
    df=dataframe.copy()
    for each in df.columns:
        print("Original Datatype :", each," : ", df[each].dtype)
        if (df[each].dtype =="object") & ( each in categorical_list):
            print("okay",each)
        elif (df[each].dtype =="object") & (each in Numerical_list):
            print("Needs Correction into Numerical :",each)
            try:
                df[each]=df[each].astype("float64")
                print("Now Converted to :",df[each].dtype)
            except:
                print("Unable to Convert directly into Numerical :", each)
                df[each]=remove_wrong_entries(df[each],float)
                df[each]=df[each].astype("float64")
                print("Now Converted to :",df[each].dtype)
        elif (df[each].dtype !="object") & ( each in categorical_list):
            df[each]=df[each].astype("object")
            print("Now Converted to :",df[each].dtype)
        else:
            print("Everything is fine")
        print("\n")
    print("\n")
    return df
#############################################################            
#############################################################                  
            
data1=mf(clean_dataframe(data1,Categorical_columns,Numerical_columns))
```

```python
###########################################################
############################################################
#Function that will list out column by column outliers' Index values. We will check patterns for those.
def outlier_identification(dataframe):
    outliers_values={}
    dataframe_desc=dataframe.describe()
    for col in dataframe.columns:
        Quartile3=list(dataframe[col][dataframe[col]>dataframe_desc[col]["75%"]].index)
        Quartile1=list(dataframe[col][dataframe[col]<dataframe_desc[col]["25%"]].index)
        outliers_values.update({col:[Quartile1,Quartile3]})
    return outliers_values
###########################################################
############################################################
outliers_default=pd.DataFrame(outlier_identification(data1_default_wo[Numerical_columns]),index=["Q1","Q3"])
#outliers_default


outliers_nodefault=pd.DataFrame(outlier_identification(data1_nodefault_wo[Numerical_columns]),index=["Q1","Q3"])
#outliers_nodefault
```

```python
###########################################################
###########################################################
def append_data_within(dataframe,function=len):
    k=[]
    for row in dataframe.index:
        temp=[]
        for col in dataframe.columns:
            temp.append(function(dataframe[col][row]))
        #temp.append(temp[0]+temp[1])
        k.append(temp)
    return k
#############################################################
#############################################################
outliers_nodefault.append(pd.DataFrame(append_data_within(outliers_nodefault),columns=list(outliers_nodefault.columns),index=["Q1","Q3"])).T
outliers_default.append(pd.DataFrame(append_data_within(outliers_default),columns=list(outliers_default.columns),index=["Q1","Q3"])).T
```

```python
####################################################################
def NaN_treatment(dataframe,Categorical_list, Numerical_list):
    df=dataframe
    j=0
    k=0
    for col in df.columns:
        if col in Numerical_list :
            mean=df[col].mean()
            #print(" Column is ",df[col].dtypes,mean)
            df[col].fillna(value=mean,axis=0,inplace=True)
            k+=1
        elif col in Categorical_list:
            mode=df[col].mode(dropna=True)[0]
            #print(mode)
            #print(" Column is ",df[col].dtypes,mode)
            df[col].fillna(value=mode,axis=0,inplace=True)
            j+=1
        print("\n")
    print("The number of categorical columns is :", j)
    print("The Number of Numerical columns is :",k)
    return df
#####################################################################
```

```python
#########################################################################
#
# This takes a list of models, generates data for them and then sends out
# parameters.
#
# Writing a function to cover the above in a pipeline if we want.
#########################################################################
#########################################################################
def Model_fitting_parameter_extraction(X_train,y_train,X_test,y_test,list_of_models):
    #################################################################
    from sklearn.metrics import accuracy_score
    from sklearn.metrics import recall_score
    from sklearn.metrics import recall_score
    from sklearn.metrics import classification_report
    from sklearn.metrics import roc_auc_score as area_under_curve
    from sklearn.metrics import roc_curve, auc
    from sklearn.metrics import confusion_matrix
    from sklearn import metrics
    ###################################################################
    comparison_of_models_matrix=[]
    classification_report_list={}
    confusion_matrix_list={}
    roc_characteristics={}
    ###################################################################
    #
    ########################################################################
    #########################################################################    
    for Model in list_of_models:
        print(type(Model))
        Model.fit(X_train,y_train)
        #print("Step 1 : Finished")
        Model_prediction=Model.predict(X_test)
        train_score=Model.score(X_train,y_train)
        test_score=Model.score(X_test,y_test)
        Accuracy=accuracy_score(y_test,Model_prediction)
        Recall=recall_score(y_test,Model_prediction)
        area=area_under_curve(y_test,Model_prediction)
        probability=Model.predict_proba(X_test)
        #print(probability.shape)
        False_p_rate,True_p_rate,thresh=roc_curve(np.array(y_test.ravel()),probability[:,1])
        #print("Step 2 : Finished")
        confusion_matrix_list.update({Model:confusion_matrix(y_test, Model_prediction, labels=[-1, 1])})
        #print("Step 3 : Finished")
        comparison_of_models_matrix.append({"Model":Model,
                                    "Train_score":train_score,
                                    "Test_score":test_score,"Accuracy":Accuracy,"Recall":Recall,"ROC_AUC":area})
        #print("Step 4 : Finished")
        #comparison_of_models_matrix.update(model_information(X_train,y_train,X_test,y_test,Model_prediction))
        #print("Step 5 : Finished")
        roc_characteristics.update({"False_psotive_rate":False_p_rate,
                                           "True_positive_rate": True_p_rate,"Threshold_rate":thresh})
        classification_report_list.update({"Model":Model,
                                   "classification_report":classification_report(y_test, Model_prediction, labels=[-1, 1])})
    ########################################################################
    return [comparison_of_models_matrix,classification_report_list,confusion_matrix_list,roc_characteristics]
########################################################################
########################################################################
```

```python
########################################################################
#########################################################################
#This function takes a list of confusion matrices from all the models and 
#gives out a nice dataframe. We can add further functions to this to create
# and add new featues which is given as a comment here. Define the new
# functions that return a pandas series given the confusion matrix, we can
# always create new ways.
##########################################################################
###########################################################################
def confusion_matrix_df(confusion_matrix_given):
    confusion_df_all=pd.DataFrame(confusion_matrix_given,columns=["TP","FP","FN","TN"])
    #print(confusion_df_all)
    true_p=confusion_df_all["TP"]
    false_p=confusion_df_all["FP"]
    false_n=confusion_df_all["FN"]
    true_n=confusion_df_all["TN"]
    confusion_df_all["Precision"]=true_p/(true_p+false_p)
    return confusion_df_all

#################################################################################
##################################################################################
```

```python
#################################################################################
##################################################################################
def complete_df_models(X_train,y_train,X_test,y_test,Models_list):
    dtta=[]
    dtta=Model_fitting_parameter_extraction(X_train,y_train,X_test,y_test,Models_list)
    confusion_matrix_list=[]
    #######################################################################
    for key in dtta[2]:
        #print(key)
        #print(dtta[2][key])
        confusion_matrix_list.append(dtta[2][key].ravel())
    #print(confusion_matrix_list)
    ####################################################################################
    Model_comparison=pd.DataFrame(dtta[0])
    #Model_comparison.drop_duplicates(subset="Model",inplace=True)
    Model_complete=Model_comparison.join(confusion_matrix_df(confusion_matrix_list))
    Model_complete["f1_score"]=2*(Model_complete["Precision"]*Model_complete["Recall"])/(Model_complete["Precision"]+Model_complete["Recall"])
    ##################################################################################
    return Model_complete#,dtta
#################################################################################
##################################################################################
```
