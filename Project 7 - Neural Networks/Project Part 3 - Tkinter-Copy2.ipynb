```python
def Neural_Netowrk_reg(dataframe_given,target_column):
    data=dataframe_given.copy()
    X_train=data.drop(target_column,axis=1)
    y_train=data.pop(target_column)
    ###################################################################
    #Regressional Neural Network
    ###################################################################
    model_reg=k.Sequential()
    model_reg.add(BatchNormalization(input_shape=(X_train.shape[1],)))
    model_reg.add(Flatten())
    model_reg.add(Dense(50,activation="relu",kernel_initializer='random_uniform',bias_initializer='zeros',kernel_regularizer="l2",bias_regularizer="l2"))
    model_reg.add(Dropout(0.2, input_shape=(50,)))
    model_reg.add(Dense(50,activation="relu",kernel_initializer='random_uniform',bias_initializer='zeros',kernel_regularizer="l2",bias_regularizer="l2"))
    model_reg.add(Dropout(0.2, input_shape=(50,)))
    model_reg.add(Dense(30,activation="relu",kernel_initializer='random_uniform',bias_initializer='zeros',kernel_regularizer="l2",bias_regularizer="l2"))
    #model_reg.add(Dropout(0.5, input_shape=(50,)))
    #model_reg.add(Dense(50,activation="relu",kernel_initializer='random_uniform',bias_initializer='zeros',kernel_regularizer="l2",bias_regularizer="l2"))
    #model_reg.add(Dropout(0.2, input_shape=(50,)))
    #model_reg.add(Dense(50,activation="relu",kernel_initializer='random_uniform',bias_initializer='zeros',kernel_regularizer="l2",bias_regularizer="l2"))
    model_reg.add(Dropout(0.5, input_shape=(30,)))
    model_reg.add(Dense(50,activation="relu",kernel_initializer='random_uniform',bias_initializer='zeros',kernel_regularizer="l2",bias_regularizer="l2"))
    model_reg.add(Dense(1))
    sgd = optimizers.SGD(lr = 0.01,momentum=0.6)
    model_reg.compile(optimizer = sgd, loss = 'mse', metrics =k.metrics.MeanSquaredError())
    ###################################################################
    #
    ###################################################################
    history_reg=model_reg.fit(X_train,y_train,batch_size=100, epochs = 500)
    return history_reg
```
