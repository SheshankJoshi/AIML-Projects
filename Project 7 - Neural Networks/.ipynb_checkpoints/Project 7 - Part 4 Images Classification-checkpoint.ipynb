```python
import tensorflow as tf
import h5py
import tensorflow.keras as k
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import img_to_array
import itertools
import numpy as np
```

```python
file=h5py.File("Part - 4 - Autonomous_Vehicles_SVHN_single_grey1.h5","r+")
```

```python
print(file)
```

```python
file.keys()
```

```python
X_test=tf.convert_to_tensor(file["X_test"])
X_train=tf.convert_to_tensor(file["X_train"])
X_val=tf.convert_to_tensor(file["X_val"])
y_test=tf.convert_to_tensor(file["y_test"])
y_train=tf.convert_to_tensor(file["y_train"])
y_val=tf.convert_to_tensor(file["y_val"])
```

```python
print("Test set Shape : ",X_test.shape,y_test.shape)
print("Train Set Shape : ",X_train.shape,y_train.shape)
print("Validation Set Shape : ",X_val.shape,y_val.shape)
```

```python
y_train=tf.reshape(y_train,shape=(y_train.shape[0],1))
print("Now the Reshaped version is : ", y_train.shape)
y_val=tf.reshape(y_val,shape=(y_val.shape[0],1))
print("Now the Reshaped version is : ", y_val.shape)
y_test=tf.reshape(y_test,shape=(y_test.shape[0],1))
print("Now the Reshaped version is : ", y_val.shape)
```

```python
#X_test = tf.reshape(X_test,shape=(X_test.shape[0], -1))
print("The Reshaped version is :", X_test.shape)
#Rest of the sets, we reshape using layers while training
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras import optimizers
from tensorflow.keras.layers import Flatten,Dropout
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Reshape
```

```python
model=Sequential()


model.add(Reshape((32*32,),input_shape=(32,32,)))

model.add(k.layers.BatchNormalization())
model.add(k.layers.Dense(60,activation='relu',kernel_initializer='he_normal',bias_initializer='zeros',kernel_regularizer="l2",bias_regularizer="l2"))
model.add(Dropout(0.2))
model.add(k.layers.Dense(60,activation='relu',kernel_initializer='he_normal',bias_initializer='zeros',kernel_regularizer="l2",bias_regularizer="l2"))
model.add(Dropout(0.4))
model.add(k.layers.Dense(60,activation='relu',kernel_initializer='he_normal',bias_initializer='zeros',kernel_regularizer="l2",bias_regularizer="l2"))
model.add(Dropout(0.4))
model.add(k.layers.Dense(60,activation="relu",kernel_initializer='he_normal',bias_initializer='zeros',kernel_regularizer="l2",bias_regularizer="l2"))
model.add(Dropout(0.2))
model.add(k.layers.Dense(10,activation='softmax'))

model.compile(optimizer='sgd', loss='categorical_crossentropy',  metrics = ['accuracy'])

history_model=model.fit(X_train,to_categorical(y_train),epochs=2000,validation_data=(X_val,to_categorical(y_val)),batch_size=150)
```

```python
y_pred=model.predict(X_test)
```

```python
###################################################################
# Function Converting predicted probabilities into class labels.
###################################################################

def convert_to_class_labels(y_predicted):
    class_predicted=[]
    y_pred=y_predicted
    class_labels=y_pred.shape[1]
    for each in y_pred :
        array=each
        for n in range(len(array)):
            if array[n]==array.max():
                class_predicted.append(n)
    return class_predicted
###################################################################
#
###################################################################
convert_to_class_labels(y_pred)
                
```

```python
fig=plt.figure(figsize=(12,12))
for i,j in itertools.combinations(range(1,6),2):
    print(i,j)
    print("The Class label is : ",convert_to_class_labels(y_pred)[i])
    ax=fig.add_subplot(i,j,i+1)
    plt.imshow(X_test[i])
    plt.show()
# These are the class labels based on the maximum value of the prediction i.e. maximum probability.
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
import seaborn as sns
print("The Testing Accuracy of the model is : ",accuracy_score(y_test.numpy(),convert_to_class_labels(y_pred)))
plt.figure(figsize=(20,12))
sns.heatmap(confusion_matrix(y_test.numpy(),convert_to_class_labels(y_pred)),annot=True)
plt.title("y_test vs y_pred")
plt.xlabel("Original")
plt.ylabel("Predicted")
plt.show()
```

**Observations**

If you observe carefully, the highest shaded is the most confused one. It is quite common that the correlations are huge between the predicted ones and the original ones if they look similar. For e.g. 
Note that there is a minimum conflict on the actual and predicted value between the value 2 and 5 as they don't look siimlar at all. On the other hand, the highest confusion is between 8 and 6 which is pretty much expected as we humans fall for it it.

Another notice is that there is no confusion between 6 and 7 but there is a huge confusion between 2 and 7. Perhaps Convolution of the samples might make this model better.
