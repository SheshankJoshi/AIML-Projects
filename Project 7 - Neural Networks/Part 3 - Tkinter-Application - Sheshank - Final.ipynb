```python
# importing tkinter and tkinter.ttk
# and all their functions and classes
from tkinter import * 
from tkinter.ttk import *
import tkinter as tk
########################################################################################       
########################################################################################
#####
########################################################################################
########################################################################################
class MyWindow(tk.Frame):
    # importing askopenfile function
    # from class filedialog
    from tkinter.filedialog import askopenfile
    from tkinter.filedialog import askopenfilename
    from tkinter.messagebox import showinfo
    from tkinter.messagebox import askquestion
    import tkinter as tk
    import numpy as np
    import pandas as pd
    import scipy as sp
    import itertools
    import time
    #import warnings
    #warnings.simplefilter("ignore")
    from tensorflow.keras.layers import Dense
    from tensorflow.keras import regularizers, optimizers
    from tensorflow.keras.models import Sequential
    from tensorflow.keras.utils import to_categorical
    from tensorflow.keras.layers import BatchNormalization, Dropout,Reshape,Activation,Dense,Flatten
    ###############################################
    file_retrived="File Imported Successfully"
    file_not_selected="Please select proper file"
    target_column_found="Target Column Found Successfully"
    target_column_invalid="Target Column Invalid"
    network_trained="Neural Network Trained Successfully"
    network_not_trained="Neural Network Not Trained"
    file_save_success="Network Pickled Successfully"
    file_save_failure="Unknown Failure while pickling"
    ###############################################
    #
    def __init__(self,master):
        super().__init__(master)
        ###################################################################
        # Object related Static Data
        ################################################################### 
        #Frame(self)
        self.open=True
        self.master=master
        master.title("Neural Network - GUI Interface")
        #print("Type of master is : ",type(self.master))
        #print("self is : ",type(self))
        #for item in dir(self):
            #print (item,"\n")
        self.message1=""
        self.message2=""
        self.message3_1=""
        self.message4_1=""
        self.message3_2=""
        self.message4_2=""
        self.df_out=None  # Data from the file stored as a pandas dataframe
        self.file=tk.StringVar() # File string of the exact location
        self.target_col=tk.StringVar() # Target column variable
        self.not_chosen_a_file=True
        self.opened_file_name=""
        self.model_out_cat=None # Our Original Categorical Model
        self.model_out_reg=None # Our Original Regression Model
        self.progress_reg=False
        self.progress_cat=False
        #frame1=Frame(self)
        #frame1.pack(side=TOP,fill=X)
        #print("Type of frame is : ",type(frame1))
        
        ###################################################################
        # End of Static Data Initialization
        ################################################################### 
        
        ###################################################################
        # Frame Initializations
        ################################################################### 
        
        self.frame1=self.frame_setup()
        self.frame1.pack(side=TOP,fill=X)
        #
        self.frame2=self.frame_setup()
        self.frame2.pack(side=TOP,fill=X)
        #
        self.frame3=self.frame_setup()
        self.frame3.pack(side=TOP,fill=X)
        #
        self.frame4=self.frame_setup()
        self.frame4.pack(side=TOP,fill=X)
        #
        ###################################################################
        # End of Frame Initializations
        ###################################################################
        
        self.check=False # Just a check variable.
        
        ###################################################################
        # Column grid Initializations
        ################################################################### 
        self.Column1_init()
        self.Column2_init()
        self.Column3_init()
        self.Column4_init()
        ###################################################################
        # End of Column Grid Initializations
        ################################################################### 
        
        self.open=False # Another Check Variable
        #self.labelframe.mainloop()
        #self.mainloop()
    
    ###################################################################
    # Frame Setup Method
    ################################################################### 
    
    def frame_setup(self):
        return Frame(self.master)
    
    ###################################################################
    # End of Frame Setup Method
    ################################################################### 
    
    ###################################################################
    # Column1 Initialization Method
    ################################################################### 
    
    def Column1_init(self):

        w_width=len("Step 3 : Neural Network Regressor")
        label1=Label(self.frame1,text="Step 1 : Import File",justify=LEFT,anchor=W,width=w_width)
        label1.grid(row=0,column=0,sticky=N+W+E+S,padx=10)
        #
        label2=Label(self.frame2,text="Step 2 : Specify Target Column",justify=LEFT,anchor=W,width=w_width)
        label2.grid(row=0,column=0,padx=10)
        #
        label3=Label(self.frame3,text="Step 3 : Neural Network Regressor",justify=LEFT,anchor=W,width=w_width)
        label3.grid(row=0,column=0,padx=10)
        #
        label4=Label(self.frame3,text="Regression",justify=RIGHT,anchor=E,width=w_width)
        label4.grid(row=1,column=0,padx=10)
        #
        label5=Label(self.frame3,text="Pickle",justify=RIGHT,anchor=E,width=w_width)
        label5.grid(row=2,column=0,padx=10)
        #
        label6=Label(self.frame4,text="Step 4 : Neural Network Classifier",justify=LEFT,anchor=W,width=w_width)
        label6.grid(row=0,column=0,padx=10)
        #
        label7=Label(self.frame4,text="Classifier",justify=RIGHT,anchor=E,width=w_width)
        label7.grid(row=1,column=0,padx=10)
        #
        label12=Label(self.frame4,text="Pickle",justify=RIGHT,anchor=E,width=w_width)
        label12.grid(row=2,column=0,padx=10)
        #
        btn_refresh=Button(self.frame4,text="Refresh",command=lambda : self.refresh_button()) #should add command for refreshing
        btn_refresh.grid(row=3,column=0,columnspan=2,sticky=W,padx=10,pady=10)
    
    ###################################################################
    # End of Column1 Initialization Method
    ################################################################### 
    
    ###################################################################
    # Column2 Initialization Method
    ###################################################################     
    
    def Column2_init(self,destroy=False):
        if (self.check):
            print("Successfully re-initialized Column2")
        length_2=27
        labelframe=LabelFrame(self.frame1,text="Open file",width=length_2)
        labelframe.grid(row=0,column=1,padx=5)
        #labelframe.grid_propagate()
        if self.not_chosen_a_file:
            #if self.check:
                #print("Button Still Loading")
            btn1 = Button(labelframe, text ='Browse',command = lambda : self.openfile(),width=length_2)
            btn1.pack(fill=X)
        else:
            #if self.check:
                #print("Textframe is loading")
            textframe=Text(labelframe,height=1,width=length_2-7)
            #scrollbar = Scrollbar(textframe,orient=HORIZONTAL)
            #scrollbar.pack(side = BOTTOM )            
            #scrollbar.config(command=textframe.xview)
            #textframe.config(xscrollcommand=scrollbar.set)
            self.opened_file_name=self.file.split("/")[-1]
            textframe.insert(END,self.opened_file_name)
            textframe.pack()
        #
        labelframe2=LabelFrame(self.frame2,text="Type the target Column")
        labelframe2.grid(row=0,column=1,padx=5)
        labelframe2.grid_propagate()
        #length_2=labelframe2.winfo_width()
        #length_2=35
        #print(length_2)
        entry_target=Entry(labelframe2,textvariable=self.target_col,width=length_2)
        entry_target.pack(fill=X)
        #
        fr2=Frame(self.frame3)
        fr2.grid(row=1,column=1,padx=5)
        fr2.grid_propagate()
        btn2=Button(fr2,text="Train",width=length_2,command=lambda : self.train_model_reg())
        btn2.pack(fill=X)
        #
        fr3=Frame(self.frame3)
        fr3.grid(row=2,column=1,padx=5)
        fr3.grid_propagate()
        btn3=Button(fr3,text="Run",width=length_2,command=lambda : self.pickled_model_regression())
        btn3.pack(fill=X)
        #
        fr4=Frame(self.frame4)
        fr4.grid(row=1,column=1,padx=5)
        fr4.grid_propagate()
        btn4=Button(fr4,text="Train",width=length_2,command=lambda : self.train_model_cat())
        btn4.pack(fill=X)
        #
        fr5=Frame(self.frame4)
        fr5.grid(row=2,column=1,padx=5)
        fr5.grid_propagate()
        btn5=Button(fr5,text="Run",width=length_2,command=lambda : self.pickled_model_cat())
        btn5.pack()
        #
        if destroy:
            labelframe.destroy()
            labelframe2.destroy()
            fr2.destroy()
            fr3.destroy()
            fr4.destroy()
            fr5.destroy()
            

    ###################################################################
    # End of Column2 Initialization Method
    ################################################################### 
    
    ###################################################################
    # Column 3 Initialization Method
    ################################################################### 
    
    def Column3_init(self,destroy=False):
        length=len("Saved Model to the Disk")
        #
        fr6=Frame(self.frame1)
        fr6.grid(row=0,column=2,padx=5)
        fr6.grid_propagate()
        btn6=Button(fr6,text="Import Data",command=lambda : self.import_data_button(),width=length) 
        btn6.pack(fill=Y)
        #
        fr7=Frame(self.frame2)
        fr7.grid(row=0,column=2,padx=5)
        fr7.grid_propagate()
        btn7=Button(fr7,text="Import Target",command=lambda : self.target_col_selection(),width=length)
        btn7.pack(fill=Y)
        # Hide these columns/widgets until appropriate functions are done. Use the measure in the widget function.

        labelframe3=LabelFrame(self.frame3,text="Train Status")
        text1=Label(labelframe3,width=length)
        text1["text"]=self.message3_1 #have to change this later
        text1.pack()
        labelframe3.grid(row=1,column=2,padx=5)
        #
        labelframe4=LabelFrame(self.frame3,text=" Save Status")
        text2=Label(labelframe4,width=length)
        text2["text"]=self.message4_1 # have to change this later
        text2.pack()
        labelframe4.grid(row=2,column=2,padx=5)
        #
        labelframe5=LabelFrame(self.frame4,text="Train Status")
        text3=Label(labelframe5,width=length)
        text3["text"]=self.message3_2# We have to change this later
        text3.pack()
        labelframe5.grid(row=1,column=2,padx=5)
        #
        labelframe6=LabelFrame(self.frame4,text="Save Status")
        text4=Label(labelframe6,width=length)
        text4["text"]=self.message4_2# We have to change this later
        text4.pack()
        labelframe6.grid(row=2,column=2,padx=5)
        if destroy:
            fr6.destroy()
            fr7.destroy()
            labelframe3.destroy()
            labelframe4.destroy()
            labelframe5.destroy()
            labelframe5.destroy()
            labelframe6.destroy()
    
    
    ###################################################################
    # End of Column3 Initialization Method
    ################################################################### 
    
    ###################################################################
    # Column 4 Initialization Method
    ###################################################################    
    
    def Column4_init(self,destroy=False):
        #
        text_success="Open File Success at Location "
        labelframe7=LabelFrame(self.frame1,text="Open File Status")
        labelframe7.grid(row=0,column=3,padx=5)
        text5=Label(labelframe7,width=len(text_success))
        text5["text"]=self.message1
        text5.pack()
        if self.open:
            labelframe7.grid_forget()
        #
        fr8=Frame(self.frame2,width=len(text_success))
        fr8.grid(row=0,column=3,padx=5)
        text6=Label(fr8)
        text6["text"]=self.message2
        text6.pack(fill=X)
        if self.open:
            fr8.grid_forget()
        #
        btn_Exit=Button(self.frame4,text="Exit",command=lambda : self.exit_button())
        btn_Exit.grid(row=3,column=2,columnspan=2,sticky=E,pady=10)
        #
        text7=Label(self.frame4)
        text7["text"]="Designed by Sheshank Joshi for Great Learning Project"
        text7.grid(row=4,columnspan=4,sticky=W)
        #
        #print("properly Re-entering the picture")
        from tkinter.ttk import Progressbar
        bar_training_reg = Progressbar(self.frame3, style='grey.Horizontal.TProgressbar',mode="indeterminate")
        bar_training_reg['value'] = 100
        bar_training_reg.grid(column=3,columnspan=2, row=1)
        if not(self.progress_reg):
            bar_training_reg.grid_forget()
        if self.progress_reg:
            bar_training_reg.start()
            bar_training_reg.grid()
        else:
            bar_training_reg.grid_forget()
        #
        print("properly Re-entering the picture")
        bar_training_cat = Progressbar(self.frame4, style='grey.Horizontal.TProgressbar',mode="indeterminate")
        bar_training_cat['value'] = 100
        bar_training_cat.grid(column=3,columnspan=2, row=1)
        if not(self.progress_cat):
            bar_training_cat.grid_forget()
        if self.progress_cat:
            bar_training_cat.start()
            bar_training_cat.grid()
        else:
            bar_training_cat.grid_forget()
        if destroy:
            bar_training_reg.destroy()
            bar_training_cat.destroy()
            btn_Exit.destroy()
            fr8.destroy()
            labelframe7.destroy()
        #
    ###################################################################
    # End of Column4 Initialization Method
    ################################################################### 
        
    ###################################################################
    # Import Data Button Trigger handler method.
    ################################################################### 
    
    def import_data_button(self):
        import pandas as pd
        print("Entered the import data function")
        if self.file:
            try:
                self.df_out=pd.read_csv(self.file)
                self.message1=MyWindow.file_retrived
            except :
                self.message1=MyWindow.file_not_selected
            finally:
                print("inside import data function")
                self.open=False
                self.Column3_init(destroy=True)
                self.Column3_init()
                self.Column4_init(destroy=True)
                self.Column4_init()
    
    ###################################################################
    # End of Method to handle Import Data Button
    ################################################################### 
    
    ###################################################################
    # Widget handler for Text Entry column for Target column Selection
    ###################################################################         
    
    def target_col_selection(self):
        print("target column function reached")
        if self.df_out is not None:
            cols=self.df_out.columns
            try:
                print(self.target_col.get())
                assert self.target_col.get() in cols
                self.message2=MyWindow.target_column_found
            except :
                self.message2=MyWindow.target_column_invalid
            finally:
                print("target column selected")
                self.Column3_init()
                self.Column4_init()
        else:
            self.message2="Please Select Data First"
            self.Column3_init()
    
    ###################################################################
    # End of Target Column selection Method
    ################################################################### 
    
    ###################################################################
    # Open File Method for Browse Button
    ################################################################### 
    
    def openfile(self):
        #print("file open function breached")
        if self.open:
            pass
        else:
            #print("file open function accessed")
            from tkinter.filedialog import askopenfilename
            self.file=askopenfilename()
            if self.file:
                self.not_chosen_a_file=False
                #print("File Successfully accessed",self.file)
                #self.check=True
                #print(self.file.split("/"))
                self.Column2_init()
    
    ###################################################################
    # End of Browse Button Method
    ################################################################### 
    
    ###################################################################
    # Method for Button Trigger for Categorical Training
    ################################################################### 
    
    def train_model_cat(self):
        from tkinter.messagebox import showinfo
        print("Entered the Model handler")
        self.progress_cat=True
        self.Column4_init(destroy=True)
        self.Column4_init()
        if (self.message2==MyWindow.target_column_found):
            print("next step")
            try:
                self.progress_cat=True
                self.Column4_init()
                #print("Trial Step")
                self.Neural_Network_cat()
                assert self.model_out_cat is not None
                self.message3_2=MyWindow.network_trained
            except:
                print("Error in importing the model")
                self.message3_2=MyWindow.network_not_trained
            finally:
                self.progress_cat=False
                self.Column4_init(destroy=True)
                self.Column4_init()
                self.Column3_init(destroy=True)
                self.Column3_init()
            showinfo(title="Status of the Model Training",message=self.message3_2)
    ###################################################################
    # End of Method for button Trigger for Categorical 
    ################################################################### 
    
    ###################################################################
    # Training button Trigger method for Regression
    ###################################################################            
    
    def train_model_reg(self):
        from tkinter.messagebox import showinfo
        import time
        self.progress_reg=True
        self.Column4_init(destroy=True)
        self.Column4_init()
        if (self.message2==MyWindow.target_column_found):
            try:
                self.Neural_Network_reg()
                assert self.model_out_reg is not None
                self.message3_1=MyWindow.network_trained
            except:
                print("Error in importing the model")
                self.message3_1=MyWindow.network_not_trained
            finally:
                self.progress_reg=False
                self.Column4_init(destroy=True)
                self.Column4_init()
                self.Column3_init(destroy=True)
                self.Column3_init()
            showinfo(title="Status of the Model Training",message=self.message3_1)
    
    ###################################################################
    # End of Training Button Trigger Method
    ################################################################### 
    
    ###################################################################
    # Picking Method for button Trigger for Categorical 
    ################################################################### 
    
    def pickled_model_cat(self):
        from tkinter.messagebox import showinfo
        import pickle
        #import h5py
        if(self.message3_2==MyWindow.network_trained):
            try:
                file_open=open("NN_cat"+".sav","xb") #Do proper formatting of the file string as per wish
                pickle.dump(self.model_out_cat,file_open)
                file_open.close()
                #self.model_out_cat.save("NN_cat"+".h5")
                self.message4_2=MyWindow.file_save_success
            except:
                self.message4_2=MyWindow.file_save_failure
            finally:
                showinfo(title="Status of the Model Pickling",message=self.message4_2)
    
    ###################################################################
    # End of pickling method for button Trigger method
    ################################################################### 
   
    ###################################################################
    # Method for pickling Regression Model
    ################################################################### 
    
    def pickled_model_regression(self):
        from tkinter.messagebox import showinfo
        import pickle
        #import h5py
        print(self.message3_1)
        if(self.message3_1==MyWindow.network_trained):
            try:
                file_open=open("NN_reg"+".sav","xb")
                pickle.dump(self.model_out_reg,file_open)
                file_open.close()
                #self.model_out_cat.save("NN_reg"+".h5")
                self.message4_1=MyWindow.file_save_success
            except:
                self.message4_1=MyWindow.file_save_failure
            finally:
                showinfo(title="Status of the Model Pickling",message=self.message4_1)
    
    ###################################################################
    # End of method for pickled model regression
    ################################################################### 
    
    ###################################################################
    # Refresh button function method
    ################################################################### 
    
    def refresh_button(self):
        self.Column2_init(destroy=True)
        self.open=True
        self.Column2_init(destroy=True)
        self.Column2_init()
    
    ###################################################################
    # End of Refresh function method
    ################################################################### 
    
    ###################################################################
    # Exit Button method
    ###################################################################    
    
    def exit_button(self):
        from tkinter.messagebox import askquestion
        answer=askquestion(title="You have clicked exit", message="Do You Really Want to Exit before Saving ?")
        if answer=="yes":
            import sys
            self.master.quit()
            self.master.destroy()
    
    ###################################################################
    # End of Exit Button Method
    ################################################################### 
    
    ###################################################################
    #Categorical Neural Network
    ###################################################################
    
    def Neural_Network_cat(self):
        from tensorflow.keras.layers import Dense
        from tensorflow.keras import regularizers, optimizers
        from tensorflow.keras.models import Sequential
        from tensorflow.keras.utils import to_categorical
        from tensorflow.keras.layers import BatchNormalization, Dropout,Reshape,Activation,Dense,Flatten
        import tensorflow.keras as k
        ####################################################
        print("Entered Neural Network Categorical")
        data=self.df_out.copy()
        X_train=data.drop(self.target_col.get(),axis=1)
        y_train=data.pop(self.target_col.get())
        self.model_out_cat=k.Sequential()
        #self.model_out_cat(Flatten(input_shape=(X_train.shape[1],)))
        #self.model_out_cat.add(Reshape((784,),input_shape=(X_train.shape[0],X_train.shape[1],)))
        self.model_out_cat.add(BatchNormalization())
        self.model_out_cat.add(Dense(60,activation="sigmoid",kernel_initializer='random_normal',bias_initializer='zeros',kernel_regularizer="l2",bias_regularizer="l2"))
        self.model_out_cat.add(Dropout(0.2, input_shape=(60,)))
        self.model_out_cat.add(Dense(30,activation="relu",kernel_initializer='random_normal',bias_initializer='zeros',kernel_regularizer="l2",bias_regularizer="l2"))
        self.model_out_cat.add(Dropout(0.2, input_shape=(30,)))
        #self.model_out_cat.add(Dense(60,activation="sigmoid",kernel_initializer='random_normal',bias_initializer='zeros',kernel_regularizer="l2",bias_regularizer="l2"))
        #self.model_out_cat.add(Dropout(0.4, input_shape=(60,)))
        #self.model_out_cat.add(Dense(60,activation="relu",kernel_initializer="random_normal",bias_initializer='zeros',kernel_regularizer="l2",bias_regularizer="l2"))
        #self.model_out_cat.add(Dropout(0.2, input_shape=(60,)))
        self.model_out_cat.add(Dense(60,activation="sigmoid",kernel_initializer='random_normal',bias_initializer='zeros',kernel_regularizer="l2",bias_regularizer="l2"))
        self.model_out_cat.add(Dropout(0.4, input_shape=(60,)))
        self.model_out_cat.add(Dense(30,activation="sigmoid",kernel_initializer='random_normal',bias_initializer='zeros',kernel_regularizer="l2",bias_regularizer="l2"))
        self.model_out_cat.add(Dropout(0.2, input_shape=(30,)))
        self.model_out_cat.add(Dense(30,activation="sigmoid",kernel_initializer='random_normal',bias_initializer='zeros',kernel_regularizer="l2",bias_regularizer="l2"))
        self.model_out_cat.add(Dropout(0.2, input_shape=(30,)))
        self.model_out_cat.add(Dense(15,activation="sigmoid",kernel_initializer="random_normal",bias_initializer='zeros',kernel_regularizer="l2",bias_regularizer="l2"))
        self.model_out_cat.add(Dense(9,activation="softmax"))
        sgd = optimizers.SGD(lr = 0.01,momentum=0.3)
        self.model_out_cat.compile(optimizer = sgd, loss = 'mse', metrics =k.metrics.CategoricalAccuracy())
        self.model_out_cat.fit(X_train,k.utils.to_categorical(y_train),batch_size=100, epochs = 500, verbose = 1)

    ###################################################################
    #End of Model
    ###################################################################

    ###################################################################
    #Regressional Neural Network
    ###################################################################
    def Neural_Network_reg(self):
        from tensorflow.keras.layers import Dense
        from tensorflow.keras import regularizers, optimizers
        from tensorflow.keras.models import Sequential
        from tensorflow.keras.utils import to_categorical
        from tensorflow.keras.layers import BatchNormalization, Dropout,Reshape,Activation,Dense,Flatten
        import tensorflow.keras as k
        ######################################################
        print("Entered the Neural Network Regression")
        data=self.df_out.copy()
        X_train=data.drop(self.target_col.get(),axis=1)
        y_train=data.pop(self.target_col.get())
        self.model_out_reg=k.Sequential()
        self.model_out_reg.add(BatchNormalization(input_shape=(X_train.shape[1],)))
        self.model_out_reg.add(Flatten())
        self.model_out_reg.add(Dense(50,activation="relu",kernel_initializer='random_uniform',bias_initializer='zeros',kernel_regularizer="l2",bias_regularizer="l2"))
        self.model_out_reg.add(Dropout(0.2, input_shape=(50,)))
        self.model_out_reg.add(Dense(50,activation="relu",kernel_initializer='random_uniform',bias_initializer='zeros',kernel_regularizer="l2",bias_regularizer="l2"))
        self.model_out_reg.add(Dropout(0.2, input_shape=(50,)))
        self.model_out_reg.add(Dense(30,activation="relu",kernel_initializer='random_uniform',bias_initializer='zeros',kernel_regularizer="l2",bias_regularizer="l2"))
        #self.model_out_reg.add(Dropout(0.5, input_shape=(50,)))
        #self.model_out_reg.add(Dense(50,activation="relu",kernel_initializer='random_uniform',bias_initializer='zeros',kernel_regularizer="l2",bias_regularizer="l2"))
        #self.model_out_reg.add(Dropout(0.2, input_shape=(50,)))
        #self.model_out_reg.add(Dense(50,activation="relu",kernel_initializer='random_uniform',bias_initializer='zeros',kernel_regularizer="l2",bias_regularizer="l2"))
        self.model_out_reg.add(Dropout(0.5, input_shape=(30,)))
        self.model_out_reg.add(Dense(50,activation="relu",kernel_initializer='random_uniform',bias_initializer='zeros',kernel_regularizer="l2",bias_regularizer="l2"))
        self.model_out_reg.add(Dense(1))
        sgd = optimizers.SGD(lr = 0.01,momentum=0.6)
        self.model_out_reg.compile(optimizer = sgd, loss = 'mse', metrics =k.metrics.MeanSquaredError())
        self.model_out_reg.fit(X_train,y_train,batch_size=100, epochs = 250)
        
    ###################################################################
    # End of Model
    ###################################################################
                       
########################################################################################       
########################################################################################
#####
########################################################################################
########################################################################################
if __name__=="__main__":
    root=Tk()
    root.resizable(width=False,height=False)
    root.geometry("800x350")
    window=MyWindow(root)
    window.mainloop()
```
