```python
import multiprocessing
import time
############################################################
#############################################################
def mf(function):
    import time
    multiprocessing_func=function
    starttime = time.time()
    if __name__ == '__main__':
        processes = []
        for i in range(1,10):
            p = multiprocessing.Process(target=multiprocessing_func, args=(i,))
            processes.append(p)
            p.start()    
        for process in processes:
            process.join()
        
    #print()    
    print('Time taken = {} seconds\n'.format(time.time() - starttime))
    return multiprocessing_func
#######################################################################
#####################################################################
    
#multiprocessing_func(function) as mf(function) # function() is the function we want multiprocessed
```

```python
#Demonstration of the above function
#######################################################################
#######################################################################
def finding_primes_below(n,var=False):
    if var : starttime=time.time()
    j=2
    k1=True
    k2=True
    for i in range(n-1,-1,-1):
        if k1:
            if n%j==0:
                print("The lowest divisor is :",j)
                k1=False
            j+=1
        if k2:
            if n%i==0:
                print("The highest divisor is : ",i)
                k2=False
        if not(k1 or k2):
            if var : print('Time taken = {} seconds\n'.format(time.time() - starttime))
            break
    print("Successful Job Completion : No Multiprocessing Used\n") if var else print("Successful Job Completion : Multiprocessing Used\n")
    return j-1 
#######################################################################
#######################################################################
k=mf(finding_primes_below(7879741))
print(k)
finding_primes_below(7879741,var=True)
```

# Assumptions and Pre-positions

Before we proceed we have some very strong assumptions to make regarding the data so that model might stay relevant.
1. Our Data prediction model is supposed to model only this manufacturing, hence these models can only be used for prediction of this particular product only. For different products, even using the same sensors, we might need different machine learning models.
2. Data corresponds to only type of manufacturing product and not different types and machines are always operated appropriately.
3. There might be some problem with Sensors or readings or both or none. Exact problem can't be identified unless we have domain knowledge and specified on the ground understanding of what these sensors represent.

# Strategy and Approach
**Notes** : 

From my functional knowledge of the manufacturing processes in general and the nature of the IoT devices and their connections, there are some considerations that can be made here, and the corresponding chain of thought to be explored for a given data. We can't be sure any single idea or a combination of all of them might lead to failure case.
1. Sensors often loose calibration which might drop them dead and lead them to not report values at all, or even if they report they might be dogmatic or wild swings. That unusual patterns need to be checked and correspondence to the output need to be checked.
2. Sometimes Sensors are pushed beyond their resolution, meaning that sensors might not sense the subtle changes at all, individually, but collectively, they might be able to brighten our chances of detecting subtle changes. In this process there is a real chance that some sensors might outshine others in their contribution towards some kind of sum total, those need to be appropriately studied.
3. There is a realistic chance that sensors might have lost their sensitivity or resolution because of some unusual exposure to a particular datapoint, and might have taken some time to settle down to its normal value, probably resulting in erroneous reporting. These can be evidenced in the Pass case (particularly) and might have been misreported of any anamoly. Hence it becomes important to see how pass/fail are distributed in time.
4. Usually very frequent datapoint collection, sometimes too frequent might produce some kind of interference in the data of a sensors with its own previous data creating a feedback loop for a number of cycles with almost same values. But, this pattern is extremely hard to detect and can very easily be disguised (also easily be mistaken for a abnormal reading).
5. Since Failure case and Pass case and disproportinately represented, we should look for anamolies within them, in parallel, side by side manner and compare.

**Strategy** :

1. First have a look at the data and check if they form any clusters around the data point that separates Pass case and fail case. If there are any on-the-face indicators of something being a pass case and fail case.
2. Explore the options of how the manufacturing has progressed and when it happened, i.e. a kind of time series analysis and observe some pattern. (No further analysis will be undertaken here)
3. Since the manufacturing processes are usually pretty much decided beforehand of exact proportions and we are not looking at raw or wild data but at very much expanded data. I am Assuming here that manufacturing process is tuned perfectly to the precision before these sensors were set into motion. So, there is always a set expected value for each sensor no matter what. This means that not much variations are expected within the data (in Data Science terminology, the max,min values should be within acceptable range and anything other than that is to be a serious suspect).
4. Expected value is always one particular sensor value, so anything that doesn't a very high modal count, or that there more than usual number of modes (equal in number to total number of datapoints) can be taken with suspicion for cause of the failure.
5. Almost all of the readings here are continuous values, so variations are going to be subtle, hence we can ignore them, until we find no variations at all. The mode has to be compared - not in a direct way - with the mean value. If they seem to be close enough or not (how close is close enough is a BIG question that I don't know the answer for), especially in the situations of fail case and pass case separately, and note down. If they are close enough, they can easily be replaced by their corresponding modal values (a strong reason is because modal value is expected from the sensor - especially in the pass case scenario).
6. If the above doesn't workout, then we can simply replace it by average/mean value and that can be asserted to be expected value. But we have to do them separately for fail case and pass case, with their corresponding respective means and modes.
7. Then do the minmax or zscore scaler. 
8. Select a bunch of models from supervised learning and figure out through PCA or some other means to select principal components, do some Eigen Transformations, and make the data ready.
9. Do some Hyperparameter tuning, to select for the best fit parameters.
10. Once we have that we can loop it around for bootstrap sampling. We can simultaneously go for lasso or ridge cross validation to prepare the datasets for the real model. Set up appropriate functions that perform all this.
11. After all this observe data, predictive models and their performance across the various considerations and finalize the parameters, and set up a list of these in appropriate order.
12. Create a pipeline for all this in a step by step manner : a. Take and eliminate uselessdata b.scaling c.Feature Selection d. model selection.
13. Create Multiple pipelines for multiple models.
14. Create a parameters grid, model performance parameters, suggestions on which model to choose. Depending upon the chosen model, proceed further.
15. Create a pickled file and store it for future usage.
16. Check the test set data given to us and test it against it.

```python
#This provides horizontal datarange slices for individual data entry points, especially in the light that there might be some connection
#
###############################################################
################################################################
def slice_ranges(dataframe1,dataframe2,range_slices=5):
    slice_ranges=[]
    highvalue=dataframe1.shape[0]
    lowvalue=0
    rangevalue=int(highvalue/5)
    remaindervalue=highvalue%5
    for i in range(lowvalue,highvalue,rangevalue):
        if rangevalue+lowvalue>highvalue:
            slice_ranges.append([(i,rangevalue+lowvalue),[dataframe1.loc[i:],dataframe2.loc[i:]]])
        else:
            slice_ranges.append([(i,rangevalue+lowvalue),dataframe1.loc[i:i+rangevalue],dataframe2.loc[i:i+rangevalue]])
    return slice_ranges
#######################################################################
#######################################################################
```

Now that we have dealt with all possible pattern recognitions and explorations with the data, we have a strategy to follow so that we have data ready.


**Strategy** :

1. Eliminate irrelevant sensors from the both Pass case and fail case scenarios.
2. Find out the outliers and NaN values (we have already done this)
3. Define an integer threshold parameter for considering how distribution is to be dealt with as Normal distribution and what not. We will assign this as a special parameter, or data entry point, or a separate dataframe, depending upon the size of the data.
4. We define another parameter called proportion parameters (which can perhaps be tuned depending upon how many sensors and how significant each are taking the list of values of significance (scale) given for each sensor). We will give equal proportion for all sensors, but it can always be adjusted. (It can be modified such that we have "0" as a value for irrelevant sensors indicating that they are completely dropped. We don't have deeplearning mechanism to self set these values or tweak them and learn). Depending upon those proportion threshold, if the outlier proportion lie beyond that threshold for that particular sensor, we will deal with the two conditions case by case basis.

case 1. Outliers are not within the threshold proportion : In this case we check if the particular sensor is normally distributed. If it is, we replace the outlier with the Mean value. If it is not normally distributed, we will check the modal value for that particular sensor. If the highest value mode exceeds a certain proportion (to be decided based on the importance of the sensor given by the user in the previously discussed proportion parameters, but we will fix it as a parameter based on the importance of the sensor based on the previous scale). Usually if we want more precision in our manufacturing, we should have very little variations and highest modal value. In pass case scenario we will replace those values with median and in fail case scenario we will replace them with mode value.


case2. Outliers are within the threshold proportion : In this case we check if the sensor values are normally distributed, they can be replaced with corresponding threshold values (Min-max kind of way). If they are not normally distributed, it is easy to replace them with modal value.

5. After this we go for NaN values within the dataframe. For pass case scenario and fail case scenario we have different approaches. For Fail case, we will go with highest mean value because the significance of the sensor is very important in this case and hence shouldn't affect other sensors while its calculating, and for pass case scenario we have flexibility on how we can replace the values, hence we will go for modal value, because modal value is the expected value of the sensor and since it is already a pass case scenario, our replacement won't beat the learning outcome.

```python
threshold_chosen=3 #This is the master threshold value that is going to go long way
#######################################################################
#######################################################################
def outlier_NaN_treatment(dataframe,threshold): # threshold in terms of no of std deviations something is away.
    df=dataframe
    df_desc=df.describe()
    Q2=df_desc.loc["50%"]
    Q1=Q2-(threshold*df_desc.loc["std"])
    Q3=Q2+(threshold*df_desc.loc["std"])
    for each_col in df.columns:
        #key=False
        #print("\n")
        #print("Column",each_col)
        #print("Quartile 3 Threshold:",Q3.loc[each_col])
        #print("Quartile 1 Threshold:",Q1.loc[each_col])
        col_mode=df[each_col].mode()
        #print("Before Mode :",col_mode)
        col_mode=col_mode.mean() #In case if there are multiple modes, this will level it out.
        #print("Column Mode :",col_mode)
        df[each_col].fillna(value=col_mode,axis=0,inplace=True)
        ################################################################
        for each_value in df[each_col].values:
            #print(each_value,Q3)
            outlier_above=0
            if each_value>Q3[each_col]:
                outlier_above+=1
                #if any([each_value==ale for ale in list(df[each_col])]) : print("Present")
                #print("The Number of Q3 outliers",(df[each_col]==each_value).sum())
                df[each_col].replace(to_replace=each_value,value=Q3[each_col],inplace=True)
                #if any([each_value!=ale for ale in list(df[each_col])]) : print("Now Removed")
                #The above is for Quartile3 outliers
            if each_value<Q1[each_col]:
                #print("The Number of Q1 outliers",(df[each_col]==each_value).sum())
                df[each_col].replace(to_replace=each_value,value=Q1[each_col],inplace=True)
        ####################################################################   
    return df
#######################################################################
#######################################################################
t=outlier_NaN_treatment(ddt1_fail,threshold=threshold_chosen) #Fail case
r=outlier_NaN_treatment(ddt1_pass,threshold=threshold_chosen) #Pass case
```

**Note** : 
We don't have any perfect way or automatic tuning mechanisms for having best features for any given model. So, I am writing a function here that makes use of the most general Linear Regression Model to arrive at best features to be retained, as a preliminary stage, instead of any single feature.

It can always be tweaked. But, I still suggest using PCA for better way of dealing with things.

```python
############################################################################
#
#Writing a function for pipelining mechanism and automation for the same.
#function automatically suggests bets features, trying out every possible combination
#and feature. It starts with 1 and goes all the way to highest number, finds out which
# number makes maximum cumulative difference, and suggests it.
#
############################################################################
###########################################################################
def sequential_feature_selector(X_train,y_train):
    from mlxtend.feature_selection import SequentialFeatureSelector as sfs
    from sklearn.linear_model import LinearRegression
    from sklearn.metrics import r2_score
    ######################################################################
    choice=len(X_train.columns)
    sfs1 = sfs(LinearRegression(), k_features=choice, forward=True, scoring='r2', cv=5)
    # Fitting into the Function
    sfs1=sfs1.fit(X_train.values, y_train.values)
    #print(sfs1.get_metric_dict())
    ###########################################################################
    def select_importance(dataframe,metric="avg_score"):
        minimum_threshold=0.7 #We want features to explain atleast 70 % of data
        df=dataframe.T
        l=df[df[metric]>minimum_threshold][metric]
        ind=l.diff()[l.diff()==l.diff().max()].index
        k=list(df["feature_names"].loc[ind])
        #print(k)
        return list(k[0])
    ############################################################################
    #print(pd.DataFrame(sfs1.get_metric_dict()))
    selected_columns=select_importance(pd.DataFrame(sfs1.get_metric_dict()))
    #print(selected_columns)
    columns_to_drop=[col for col in X_train.columns if col not in selected_columns]
    return X_train[selected_columns],y_train
######################################################################################################
########################################################################################
```

```python
################################################################
#This is the function we have written so that by default it takes 
# no any number of components, though if specified by user it abides by it.
# It auto-tunes itself to choose minimum number of features with maximum cumulate
# (or rate change) per increase in the number of features. Method is not rigorous
# but the essense is captured in terms of differences (should be the gradient methods).
# Such differential threshold is given in terms of minimum_threshold variable. We can tune it to purpose it.
#
#############################################################
#############################################################
def PCA_transform_features(X_train,y_train,no_of_components=None):
    ##########################################################
    from sklearn.decomposition import PCA
    #########################################################
    minimum_threshold=0.005 # We can tune this later for whatever value we want for components that can be neglected for their influence
    n=0 #Initialize the number of features to be 0.
###############################################################
    def parameter_tuning(X_train,threshold):
        parameters_to_choose=range(len(X_train.columns))
        decision_maker={}
    ####################################
        for i in parameters_to_choose:
            pca=PCA(n_components=i)
            pca.fit(X_train)
            pca_variance=pca.explained_variance_ratio_.sum()
            decision_maker.update({i:pca_variance})
    ####################################
            #print(decision_maker)
        decision_maker=pd.Series(decision_maker).diff()
        out=decision_maker[decision_maker>threshold].index.max()
        return out
#############################################################
    #print(no_of_components)
    if no_of_components==None:
        n=parameter_tuning(X_train,threshold=minimum_threshold)
        #print("Here in no components")
        #print(n)
    else:
        n=no_of_components
    pca=PCA(n_components=n,random_state=random_state)
    pca.fit(X_train)
    #print(pca.explained_variance_ratio_)
    X_trans=pd.DataFrame(pca.transform(X_train))
    return X_trans,y_train
#############################################################
############################################################# 
#completely transformed dataframe is taken. 
#We are going it real here and taking on the original data.
X_train,y_train=mf(PCA_transform_features(X_train,y_train))
#x,y=mf(PCA_transform_features(x,y)) Master dataframe Transformation.
```

```python

##########################################################################
# Function version of the same as above but with more dynamic and pipeline
# Here Synthetic generation is done via LogisticRegression as a default one, but if the user
# wants, he can tune it a lot more. Model, for sampling can be Decision Tree or anything and its
# fit into it and the score will be considered for best amount of the neighbors for the model. I am not
# sure it is the right way though. However threshold of 0.7 as a minimum chosen for any number of neighbors to 
# cross.
#
#############################################################################
##########################################################################
def SMOTE_sampling(X_train,y_train,Model=None):
    #######################################################
    #from sklearn.metrics import accuracy_score
    from imblearn.over_sampling import SMOTE
    #######################################################3
    counts=y_train.value_counts().index # We can use a for loop to extend this even further to include more than just a binary classification
    X_train_pass=y_train[y_train==counts[0]].index
    X_train_fail=y_train[y_train==counts[1]].index
    #######################################################3
    #######################################################
    if Model==None:
        from sklearn.linear_model import LogisticRegression
        Model=LogisticRegression()
    ######################################################
    ######################################################
    def Neighbor_given_out(X_train,y_train,Model):
        measurement={}
        threshold=0.7 # We can set this threshold depending upon the situation
        ############################################
        for i in range(int(X_train.shape[0]/4)): # Maximum limit we place on the number of neighbors
            sm = SMOTE(sampling_strategy = 1 ,k_neighbors = i)   #Synthetic Minority Over Sampling Technique
            print(X_train.shape,y_train.shape)
            X_train_res, y_train_res = sm.fit_resample(X_train, y_train)
            Model.fit(X_train_res,y_train_res) # Checking if the resampled versions stand against original ones
            measurement.update({i:Model.score(X_train,y_train)}) # this is a little clusmy to check against original data itself
        #############################################
        measurement=pd.Series(measurement).diff()
        out=measurement[(measurement>threshold) & (measurement<1)].index.max()
        return out
    #######################################################
    #######################################################
    neighbors_given=Neighbor_given_out(X_train,y_train,Model) # we can change this. Look at description.
    sm=SMOTE(sampling_strategy = 1 ,k_neighbors = neighbors_given)
    X_train_res, y_train_res = sm.fit_resample(X_train, y_train)
    return X_train_res,y_train_res
################################################################
#################################################################
#t_train,r_train=mf(SMOTE_sampling(X_train,y_train))
```

```python
#########################################################################
#
# This takes a list of models, generates data for them and then sends out
# parameters.
#
# Writing a function to cover the above in a pipeline if we want.
#########################################################################
#########################################################################
def Model_fitting_parameter_extraction(X_train,y_train,X_test,y_test,list_of_models):
    #################################################################
    from sklearn.metrics import accuracy_score
    from sklearn.metrics import recall_score
    from sklearn.metrics import recall_score
    from sklearn.metrics import classification_report
    from sklearn.metrics import roc_auc_score as area_under_curve
    from sklearn.metrics import roc_curve, auc
    ###################################################################
    comparison_of_models_matrix=[]
    classification_report_list={}
    confusion_matrix_list={}
    roc_characteristics={}
    ###################################################################
    #
    ########################################################################
    #########################################################################    
    for Model in list_of_models:
        Model.fit(X_train,y_train)
        #print("Step 1 : Finished")
        Model_prediction=Model.predict(X_test)
        train_score=Model.score(X_train,y_train)
        test_score=Model.score(X_test,y_test)
        Accuracy=accuracy_score(y_test,Model_prediction)
        Recall=recall_score(y_test,Model_prediction)
        area=area_under_curve(y_test,Model_prediction)
        probability=Model.predict_proba(X_test)
        #print(probability.shape)
        False_p_rate,True_p_rate,thresh=roc_curve(np.array(y_test.ravel()),probability[:,1])
        #print("Step 2 : Finished")
        confusion_matrix_list.update({Model:metrics.confusion_matrix(y_test, Model_prediction, labels=[-1, 1])})
        #print("Step 3 : Finished")
        comparison_of_models_matrix.append({"Model":Model,
                                    "Train_score":train_score,
                                    "Test_score":test_score,"Accuracy":Accuracy,"Recall":Recall,"ROC_AUC":area})
        #print("Step 4 : Finished")
        #comparison_of_models_matrix.update(model_information(X_train,y_train,X_test,y_test,Model_prediction))
        #print("Step 5 : Finished")
        roc_characteristics.update({"False_psotive_rate":False_p_rate,
                                           "True_positive_rate": True_p_rate,"Threshold_rate":thresh})
        classification_report_list.update({"Model":Model,
                                   "classification_report":classification_report(y_test, Model_prediction, labels=[-1, 1])})
    ########################################################################
    return [comparison_of_models_matrix,classification_report_list,confusion_matrix_list,roc_characteristics]
########################################################################
########################################################################
```

```python
########################################################################
#########################################################################
#This function takes a list of confusion matrices from all the models and 
#gives out a nice dataframe. We can add further functions to this to create
# and add new featues which is given as a comment here. Define the new
# functions that return a pandas series given the confusion matrix, we can
# always create new ways.
##########################################################################
###########################################################################
def confusion_matrix_df(confusion_matrix_given):
    confusion_df_all=pd.DataFrame(confusion_matrix_given,columns=["TP","FP","FN","TN"])
    #print(confusion_df_all)
    true_p=confusion_df_all["TP"]
    false_p=confusion_df_all["FP"]
    false_n=confusion_df_all["FN"]
    true_n=confusion_df_all["TN"]
    confusion_df_all["Precision"]=true_p/(true_p+false_p)
    return confusion_df_all

#################################################################################
##################################################################################
#confusion_matrix_taken=[confusion_matrix_list[x] for x in confusion_matrix_list]
####################################################################################

#Model_comparison=pd.DataFrame(comparison_of_models_matrix,index=["Model"])
#Model_comparison.drop_duplicates(subset="Model",inplace=True)
#Model_complete=Model_comparison.join(confusion_matrix_df(confusion_matrix_taken))
##################################################################################
# We need to have more than one model to have a good dataframe of comparison.
# This is built for that purpose.
#Model_complete
```

```python
#################################################################################
##################################################################################
def complete_df_models(X_train,y_train,X_test,y_test,Models_list):
    dtta=[]
    dtta=Model_fitting_parameter_extraction(X_train,y_train,X_test,y_test,Models_list)
    confusion_matrix_list=[]
    #######################################################################
    for key in dtta[2]:
        #print(key)
        #print(dtta[2][key])
        confusion_matrix_list.append(dtta[2][key].ravel())
    #print(confusion_matrix_list)
    ####################################################################################
    Model_comparison=pd.DataFrame(dtta[0])
    #Model_comparison.drop_duplicates(subset="Model",inplace=True)
    Model_complete=Model_comparison.join(confusion_matrix_df(confusion_matrix_list))
    Model_complete["f1_score"]=2*(Model_complete["Precision"]*Model_complete["Recall"])/(Model_complete["Precision"]+Model_complete["Recall"])
    ##################################################################################
    return Model_complete#,dtta
#################################################################################
##################################################################################
```

```python
# Here the no of bootstrap samples is defaulted to 1000 but more increase in 
# number is recommeneded, for more variety and confidence in our model.
########################################################################
########################################################################
def Bootstrap_confidence(data,model_given,n_iterations=100):
    ###################################################################
    from sklearn.utils import resample
    ######################################################################
    Model_chosen_final=model_given
    data1_final=data
    n_size = int(len(data1_final) * 0.50)    # picking only 50 % of the given data in every bootstrap sample
    stats = []
    in_values=data1_final.values # The Final data, that is combined and doesn't have test or train
    ######################################################################
    #print(type(in_values))
    for i in range(n_iterations):
        # prepare train and test sets
        train = resample(in_values, n_samples=n_size)  # Sampling with replacement 
        test = np.array([x for x in in_values if x.tolist() not in train.tolist()])  # picking rest of the data not considered in sample
        x_train_final=train[:,:-1]
        y_train_final=train[:,-1]
        x_test_final=test[:,:-1]
        y_test_final=test[:,-1]
        model=Model_chosen_final
        model.fit(x_train_final,y_train_final )
        # evaluate model
        predictions = model.predict(x_test_final)
        score = accuracy_score(y_test_final, predictions)    # caution, overall accuracy score can mislead when classes are imbalanced
        #print(score)
        stats.append(score)
    return stats
#######################################################################
########################################################################

stats=mf(Bootstrap_confidence(data1_final,random_forest))
```

```python
# plotting these
plt.hist(stats)
plt.show()
# confidence intervals
alpha = 0.95                             # for 95% confidence 
p = ((1.0-alpha)/2.0) * 100              # tail regions on right and left .25 on each side indicated by P value (border)
lower = max(0.0, np.percentile(stats, p))  
p = (alpha+((1.0-alpha)/2.0)) * 100
upper = min(1.0, np.percentile(stats, p))
print('%.1f confidence interval %.1f%% and %.1f%%' % (alpha*100, lower*100, upper*100))
```
