It is to be noted that phones that are sold should be sold by a minimum amount of the total phones sold to be even eligible for consideration of the top recommendations. Hence, we will eliminate all the phones that are sold just one or two in quantity and consider the ratings for the most sold phones and then sort those phones in accordance to their mean rating.

I have set here threshold to be 10% of the most sold phone's frequency (the total sales to be cut off and hence has been noted down.) from the very strong assumption that most frequently occurring product is the product that is most sold, which might be extremely wrong assumption.( I myself haven't rated a lot of items I bought).

```python
threshold_phones=value_count_products[0]/10 # This threshold value can be changed.
ratings_phones={}
######################################################################
for each_phone in value_count_products.keys():
    #print(each_phone)
    if value_count_products[each_phone]<threshold_phones:
        #print(value_count_products[each_phone])
        #print(each_phone)
        continue
    else:
        #print(each_phone)
        #print(data_stage3_product_grouping.get_group(each_phone)["score"].mean())
        ratings_phones.update({each_phone: data_stage3_product_grouping.get_group(each_phone)["score"].mean()})
####################################################################
```

```python
 # Sorting the values so that recommendations come up automatically.
####################################################################
# Function that sorts the dictionary based on the values
####################################################################
def sort_dict_values(given_dict,reverse=True):
    sorted_values=sorted(given_dict.values(),reverse=reverse)
    sorted_ratings={}
    for i in sorted_values:
        for k in given_dict.keys():
            if given_dict[k] == i:
                sorted_ratings[k]=given_dict[k]
                break
    return sorted_ratings
###############################################################
# End of funtion
###############################################################
sorted_ratings_phones = sort_dict_values(ratings_phones)
```

The above cell is copied from the internet at address given below :
https://stackabuse.com/how-to-sort-dictionary-by-value-in-python/

```python
###########################################################################
# Gives top phones based on the recommendations and the threshold as set before.
###########################################################################
def give_top_phones(sorted_ratings_phones,how_many_phones=10,min_rating=None):
    phones=[]
    if min_rating!=None:
        for each_item in sorted_ratings_phones:
            if sorted_ratings_phones[each_item]>=min_rating:
                phones.append([each_item,sorted_ratings_phones[each_item]])
        #print(phones)
        phones=pd.DataFrame(phones,columns=["Product","Avg_Rating"])
    else:
        phones=pd.DataFrame(phones,columns=["Product","Avg_Rating"])
    return phones[:how_many_phones]#pd.DataFrame(phones,columns=["Phone","Rating"])
###########################################################################
# End of function
###########################################################################
```

```python
##########################################################################################
# This function eliminates the older entries for same users same product but different times. But it should have a datetime column.
###########################################################################################

def update_user_entries(dataframe_given,columns_of_interest=["author","product"]):
    temp_index=dataframe_given[dataframe_given.duplicated(subset=columns_of_interest,keep=False)]
    indices_to_drop=list(temp_index.index)
    temp4=temp_index.groupby(columns_of_interest[0])
    for each in temp4.indices.keys():
        #print(temp_index.index.is_unique)
        #print(each)
        temp2 = temp4.get_group(each)
        temp5 = temp2.groupby(columns_of_interest[1])
        #print(temp2,temp5.indices)
        for each_product in temp2["product"].unique():
            #print(each_product)
            temp3 = temp5.get_group(each_product)
            #print(temp3)
            to_not_drop_index=temp3.date.sort_values(ascending=False).index[0]
            print(to_not_drop_index)
            indices_to_drop.remove(to_not_drop_index)
    return list(set(indices_to_drop))
##########################################################################################
#
###########################################################################################
to_drop_entries_cols=update_user_entries(data_stage7)
```

Our earlier functions gave us one list (to_drop_entries_cols) that has entries that repeated in some way or the other and the other list that contains the entire original entries as specified (i.e. more than 50 times product mentioned and more than 50 review users).


**Strategy**

Convert the users whoever are there and cap it off at 50. Let me explain as follows : so suppose if we have a particular user within these entries where the users have hundreds of entries like that of "Cliente Amazon" or "Amazon Customer"; for our practical purposes we will divide those particular entries, pick random 50 and give it a modified name like "Amazon customer 1" and "Amazon Customer2" etc. and change that for every 50 entries. So, 50 entries will be the maximum any generic user is going to get. This way, we can't loose information for our machine learning model. We will suppose that we have generated random user out of the confused and collaborated lot that we got here.

```python
kill=data_stage7_temp[data_stage7_temp["extract"].duplicated(keep=False)].groupby("author")
authors_to_check=[] # These authors have reviews that share the same date, probably for same product.
authors_to_drop=[] # These authors have clearly stated reviews that are dated properly can can be uniquely pulled out.
#######################################################################
for each in kill.indices.keys():
    #print(each)
    if any([data_stage7_temp[data_stage7_temp["author"]==each].date.is_unique]):
        #print(each) #also implies that date is unique for these ones.
        authors_to_check.append(each)
    else:
        authors_to_drop.append(each)
        #print(each)
        #
    #if any(kill.get_group(each).duplicated()):
        #print(each)
        #authors_to_check.append(each)
######################################################################
# This proves that entries are multiple in number. And nothing else is being done here.
```

We now have two lists from within the duplicated data lists that indicates two things : The first list (authors_to_drop) has authors have modified their review that hasn't been updated. So, we will simply drop the items where rating is more, assuming that lesser rating is the right rating given by the user. Also note that we have already dealt with the date entries i.e. which entries are older entries and which entries are later entries, by checking for date earlier.
The Second list (authors_to_check)

This means the same user for the same product have given two different extracts at two different times. That means something else too. We have to choose the ones that is the latest too. But, we still haven't droppped those duplicated columns from our original dataframe. We will do that with the help of given below function.

```python
##########################################################################################
#This function spits out the lists where the same extract is repeated and where the extract is not repeated.
###########################################################################################

def unique_extract_users(dataframe_given):
    to_keep=[]
    to_check=[]
    t1=dataframe_given.groupby("extract")
    for each in t1.indices:
        if len(t1.indices[each])>1:
            to_check.append(t1.indices[each])
        else:
            to_keep.append(*t1.indices[each])
    return to_keep,to_check
        #print(t1.indices[each])
##########################################################################################
#End of function.
###########################################################################################
```
