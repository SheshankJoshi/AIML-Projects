# Observations and Notes

â—‹ Batsmen having highest average and highest strike rate are the most desirable.

â—‹ But since the average can be a little flawed depending upon whether the batsmen have actually got the chance to bat, and if so at which position, will dictate what is the performance of that particular player.

â—‹ But in the data here that is completely missed. But, what comes close to it if we try to compare with Half centuries. So, we can create certain thresholds where there is a minimum number runs, minimum number of half centuries etc.

â—‹ We can't completely depend on the fours or sixes because it can be misleading to think that boundaries are reflective of the batsmen's contribution. 

â—‹ The most important feature i.e. The number of innings the batmen has played, which is not reflected here. So, ratio of the total number of runs with average gives us the no of innings played by the batsmen which indicates success/failure ratio in terms of averages.

â—‹ The development of K-Means is the appropriate method for the ranking of the batsmen, and the weightage needs, probably, be highest. Try to do the development of that feature.

**Further Thoughts**
There are some irrelevant features and the most important ones are boundaries because boundaries make very little sense, atleast for the sake of ranking. But another important feature i.e. the number of innings the batsmen got chance to play will be of importance to decide the rank because ranking without appropriate chances given to perform is unfair.
So, adding a new feature called "innings" into the dataframe and then once again ranking the models.

**Warning** : The idea of Average for a batsmen differs significantly from that of a bowler. That idea is completely being ignored here.
