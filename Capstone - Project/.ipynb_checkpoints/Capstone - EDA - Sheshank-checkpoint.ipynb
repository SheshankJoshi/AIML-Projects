# Accident Level Analysis

```python
def plot_graphs(df_given):
    df=df_given
    plt.figure(figsize=(20,12))
    plt.subplot(3,1,1)
    sns.countplot(x=df["Employee or Third Party"])
    plt.subplot(3,1,2)
    plt.hist(df["Accident Level"],label="Accident Level")
    plt.hist(df["Potential Accident Level"],label="Potential Accident Level")
    plt.legend()
    plt.show()
    plt.figure(figsize=(20,12))
    plt.subplot(3,2,1)
    sns.countplot(x=df["Industry Sector"])
    plt.subplot(3,2,2)
    sns.countplot(x=df["Local"])
    plt.show()
    plt.figure(figsize=(20,12))
    plt.subplot(3,3,1)
    sns.kdeplot(x=df["desc_count"])
    plt.subplot(3,3,2)
    sns.countplot(x=df["Countries"])
    plt.show()
```

```python
###########################################################
# Function that converts the appropriate data type and returns back the dataframe
############################################################

def condition_df(df_given):
    df=df_given
    dictionary_ref={"I":1,"II":2,"III":3,"IV":4,"V":5}
    df["Potential Accident Level"]=df["Potential Accident Level"].replace(to_replace=dictionary_ref)
    df["Accident Level"]=df["Accident Level"].replace(to_replace=dictionary_ref)
    return df
###########################################################
#
############################################################
```

```python
dictionary_ref={"I":1,"II":2,"III":3,"IV":4,"V":5}
```

### 2.1.1 Checking Potential Accident Level

```python
temp=cat["Potential Accident Level"].replace(to_replace=dictionary_ref).astype("int")
```

```python
k=temp.value_counts().sort_index()
print("The Total number of Acutal Level 1 Incidents are : ", total_cat1)
print("The Number of incidents that are mitigated and hence the incident level remained the same ",k[1])
k-k[1] # taking off the incidents that were already mitigated
data_of_interest.update({"%_translation":k*100/total_cat1})
sns.ecdfplot(k*100/total_cat1) # Cumulative Distribution plot for Potential Accidents
k*100/total_cat1
```

**Notes**
1. In total 15% of incidents were only mitigated when first reported in Level 1. Out of all the reported ones in Level 1, the highest percentage is 28 % as given above which progressed to stage 3.
2. The weightage that is assigned to the threat levels assessed should be appropriately manged in case of Level 1 reported incidents.

3. The escalation from one potential to the other, given the level 1 reported incidents is given here in terms of this cumulative plot. If we observe that it takes a long time to progress from second step to third, and equally but a little less time to go from second step to third and very quickly escalates from third to fourth. So, based on this we have to take a decision on how the progression and weightage is given, if two predictions of equal/similar probabilities are given in our model for a given description.
